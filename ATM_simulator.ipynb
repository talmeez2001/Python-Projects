```python
#TechnoHacks Edutech Python Development Internship Task 03:




'''Create a program that simulates the all atm functionalities and operations (Check balance, Deposit, Withdraw)'''




class ATM:
    def __init__(self, balance=0):
        self.balance = balance

    def check_balance(self):
        return f"\nYour account balance is ${self.balance}"

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"\n${amount} has been deposited. Your new balance is ${self.balance}"
        else:
            return "\nInvalid amount for deposit."

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"\n${amount} has been withdrawn. Your new balance is ${self.balance}"
        elif amount > self.balance:
            return "\nInsufficient funds."
        else:
            return "\nInvalid amount for withdrawal."

def main():
    atm = ATM()  # Create an ATM instance with an initial balance of $0

    while True:
        print("\n----------------Welcome to the ATM----------------")
        print("\nOptions:")
        print("\n1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Quit")

        choice = input("\nSelect an option to proceed: ")

        if choice == "1":
            print(atm.check_balance())
        elif choice == "2":
            amount = float(input("\nEnter the deposit amount: "))
            print(atm.deposit(amount))
        elif choice == "3":
            amount = float(input("\nEnter the withdrawal amount: "))
            print(atm.withdraw(amount))
        elif choice == "4":
            print("\nThank you for using the ATM. Goodbye!")
            break
        else:
            print("\nInvalid option. Please choose a valid option (1/2/3/4).")

main()
```
